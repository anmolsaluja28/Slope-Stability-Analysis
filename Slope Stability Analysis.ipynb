```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score
```

```python
data = pd.read_excel('Data.xlsx',header = [1])
X = np.array(data.iloc[:,0:-2])
y_long = np.array(data.iloc[:,-2])
y_short = np.array(data.iloc[:,-1])
y_long = y_long.reshape(y_long.shape[0],-1)
y_short = y_short.reshape(y_short.shape[0],-1)
```

```python
#Long term FOS
X_train_long,X_test_long,y_train_long,y_test_long = train_test_split(X,y_long,test_size = 0.3,random_state = 42)

#Short term FOS
X_train_short,X_test_short,y_train_short,y_test_short = train_test_split(X,y_short,test_size = 0.3,random_state = 42)
```

<b> SVR </b>

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
parameters = [{'kernel': ['rbf'], 'gamma': [1e-4, 1e-3, 0.01, 0.1, 0.2, 0.5, 0.6, 0.9],'C': [1, 10, 100, 1000, 10000]}]
svr = GridSearchCV(SVR(epsilon = 0.01), parameters, cv = 5)
svr.fit(X_train_long, y_train_long)
means = svr.cv_results_['mean_test_score']
stds = svr.cv_results_['std_test_score']
print(svr.best_params_)
print(svr.best_score_)

#for short term
parameters = [{'kernel': ['rbf'], 'gamma': [1e-4, 1e-3, 0.01, 0.1, 0.2, 0.5, 0.6, 0.9],'C': [1, 10, 100, 1000, 10000]}]
svr_short = GridSearchCV(SVR(epsilon = 0.01), parameters, cv = 5)
svr_short.fit(X_train_short, y_train_short)
print(svr_short.best_params_)
print(svr_short.best_score_)

```

```python
from sklearn.metrics import  mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
y_pred_svr = svr.predict(X_test_long)
mae_svr = mean_absolute_error(y_test_long,y_pred_svr)
mse_svr = mean_squared_error(y_test_long,y_pred_svr)
rmse_svr = np.sqrt(mse_svr)
r2_svr = r2_score(y_test_long,y_pred_svr)
print("Mean Absolute Error - " + str(mae_svr))
print("Mean Squared Error - " + str(mse_svr))
print("Root Mean Squared Error - " + str(rmse_svr))
print("R2-squared - " + str(r2_svr))
```

```python
#for short term
y_pred_svr_short = svr_short.predict(X_test_short)
mae_svr_s = mean_absolute_error(y_test_short,y_pred_svr_short)
mse_svr_s = mean_squared_error(y_test_short,y_pred_svr_short)
rmse_svr_s = np.sqrt(mse_svr_s)
r2_svr_s = r2_score(y_test_short,y_pred_svr_short)
print("Mean Absolute Error - " + str(mae_svr_s))
print("Mean Squared Error - " + str(mse_svr_s))
print("Root Mean Squared Error - " + str(rmse_svr_s))
print("R2-squared - " + str(r2_svr_s))
```

<b> MULTIPLE LINEAR REGRESSION </b>

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train_long,y_train_long)
y_pred_mlr = reg.predict(X_test_long)
mae_mlr = mean_absolute_error(y_test_long,y_pred_mlr)
mse_mlr = mean_squared_error(y_test_long,y_pred_mlr)
rmse_mlr = np.sqrt(mse_mlr)
r2_mlr = r2_score(y_test_long,y_pred_mlr)
print("Mean Absolute Error - " + str(mae_mlr))
print("Mean Squared Error - " + str(mse_mlr))
print("Root Mean Squared Error - " + str(rmse_mlr))
print("R2-squared - " + str(r2_mlr))

print("For Short term - ")
#for short term
from sklearn.linear_model import LinearRegression
reg_short = LinearRegression()
reg_short.fit(X_train_short,y_train_short)
y_pred_mlr = reg_short.predict(X_test_short)
mae_mlr = mean_absolute_error(y_test_short,y_pred_mlr)
mse_mlr = mean_squared_error(y_test_short,y_pred_mlr)
rmse_mlr = np.sqrt(mse_mlr)
r2_mlr = r2_score(y_test_short,y_pred_mlr)
print("Mean Absolute Error - " + str(mae_mlr))
print("Mean Squared Error - " + str(mse_mlr))
print("Root Mean Squared Error - " + str(rmse_mlr))
print("R2-squared - " + str(r2_mlr))
```

<b> RANDOM FOREST IMPLEMENTATION </b>

```python
from sklearn.ensemble import RandomForestRegressor
# Number of trees in random forest
n_estimators = [x for x in range(200,300,10)]
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
# Create the random grid
parameters = {'n_estimators': n_estimators,
               'max_depth': max_depth,
               }

rf = RandomForestRegressor()
# Random search of parameters, using 5 fold cross validation, 
rf_random = GridSearchCV(rf,  parameters, cv = 5)   

# Fit the random search model
rf_random.fit(X_train_long,y_train_long)


#for short term....................................
rf_short = RandomForestRegressor()
# Random search of parameters, using 5 fold cross validation, 
rf_random_short = GridSearchCV(rf_short,  parameters, cv = 5)   

# Fit the random search model
rf_random_short.fit(X_train_short,y_train_short)
```

```python
y_pred_raf = rf_random.predict(X_test_long)
mae_raf = mean_absolute_error(y_test_long,y_pred_raf)
mse_raf = mean_squared_error(y_test_long,y_pred_raf)
rmse_raf = np.sqrt(mse_raf)
r2_raf = r2_score(y_test_long,y_pred_raf)
print("Mean Absolute Error - " + str(mae_raf))
print("Mean Squared Error - " + str(mse_raf))
print("Root Mean Squared Error - " + str(rmse_raf))
print("R2-squared - " + str(r2_raf))
```

```python
#for short term..............
y_pred_raf = rf_random_short.predict(X_test_short)
mae_raf = mean_absolute_error(y_test_short,y_pred_raf)
mse_raf = mean_squared_error(y_test_short,y_pred_raf)
rmse_raf = np.sqrt(mse_raf)
r2_raf = r2_score(y_test_short,y_pred_raf)
print("Mean Absolute Error - " + str(mae_raf))
print("Mean Squared Error - " + str(mse_raf))
print("Root Mean Squared Error - " + str(rmse_raf))
print("R2-squared - " + str(r2_raf))
```
